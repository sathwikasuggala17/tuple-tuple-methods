## tuple
* Tuples are used to store multiple items in a single variable.
* Tuple is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Set, and Dictionary, all with different qualities and usage.
* A tuple is a collection which is ordered (define order, order will not change) and unchangeable(add or remove items after the tuple has been created) & allows duplicate values(Since tuples are indexed, they can have items with the same value).
* tuples are heterogenous sequential containers.
* tupples are immutable
* Tuples are written with round brackets ().

```python
a = (1, 2.2, 3+3j, 'four', print, range)
print(a, type(a), id(a), sep = '\n')
```

```python
b = tuple(range(5))
b
```

### creating python tuples
* creation of tuple in two ways :-
* 1) open and close paranthesis() 2) by using tuple class (or) tuple() constructor

```python
# 1) open & close parenthesis
c , d = ('hello'), tuple({1, 1000, 48, 99}) # c = container
c , d # c has no substrings #d has set inside tuple class
```

```python
c , d = tuple('hello'), tuple({1, 1000, 48, 99}) # c = container
c , d # c has substrings because of tuple class.
```

```python
# 2) tuple() constructor 
ramayan = tuple(('ram', 'ravan', 'sita', 'lakshman', 'hanuman')) #note double round brackets.
print(ramayan) # tuple class has substrings
```

### count()
* Returns the frequency of occurrence/ no.of times of a specified value appears in the tuple.
* return : the count of how many times element occurs in the tuple.
* syntax : tuple.count(value)
> parameter :
value : Required. The item to search for

```python
help(tuple.count)
```

```python
n = ('malayalam')
n.count('a') # a occured 4 times
```

```python
x = (1,6,2,3,6,4,6,5,6,6)
x.count(6) # 6th element occured 5 times
```

* syntax : tuple.index(value)

```python
help(tuple.index)
```
